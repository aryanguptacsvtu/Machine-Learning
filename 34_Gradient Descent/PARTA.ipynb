```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,
                      noise=80,random_state=13)
```

```python
plt.scatter(X,y)
```

## Applying OLS [Ordinary Least Squares] :

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_    # Slope
```

```python
reg.intercept_  # Intercept
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')   # Regression Line
```

## Applying Gradient Descent :
Assuming Slope is constant , m = 78.35

Assuming the starting value for intercept , b=0

```python
y_pred =  ((78.35*X) + 0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

### Iteration 1

```python
m = 78.35
b = 0

loss_slope = -2 * (np.sum(y - m*X.ravel() - b))
loss_slope
```

```python
# Taking Learning rate = 0.1
lr = 0.1

step_size = loss_slope * lr
step_size
```

```python
# Calculating the new Intercept [bnew = bold - step_size]
b = b - step_size
b 
```

```python
y_pred =  ((78.35*X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=20.927')
plt.legend()
plt.show()
```

### Iteration 2

```python
loss_slope = -2 * (np.sum(y - m*X.ravel() - b))
loss_slope
```

```python
step_size = loss_slope * lr
step_size
```

```python
# Calculating the new Intercept [bnew = bold - step_size]
b = b - step_size
b 
```

```python
y_pred =  ((78.35*X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=25.11')
plt.legend()
plt.show()
```

### Iteration 3

```python
loss_slope = -2 * (np.sum(y - m*X.ravel() - b))
loss_slope
```

```python
step_size = loss_slope * lr
step_size
```

```python
# Calculating the new Intercept [bnew = bold - step_size]
b = b - step_size
b 
```

```python
y_pred =  ((78.35*X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=25.95')
plt.legend()
plt.show()
```

### Using LOOP :

```python
b = -100
m = 78.35
lr = 0.1

epochs = 10

for i in range(epochs):
    loss_slope = -2*(np.sum(y- m*X.ravel() - b))
    b = b - (lr*loss_slope)

    y_pred = m*X + b

    plt.plot(X,y_pred)

plt.scatter(X,y)
```
