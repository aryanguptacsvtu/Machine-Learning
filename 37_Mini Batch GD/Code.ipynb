## Creating Mini Batch Gradient Descent Class :

`Batch :` Group of Rows

```python
class MBGDRegressor:
    def __init__(self,batch_size,learning_rate=0.01,epochs=100):
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size

    def fit(self,X_train,y_train):
        # Initialization
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])

        # Updation
        for i in range(self.epochs):
            for j in range(int(X_train.shape[0]/self.batch_size)):
                
                idx = random.sample(range(X_train.shape[0]),self.batch_size)  # Random numbers
                y_hat = np.dot(X_train[idx],self.coef_) + self.intercept_

                intercept_der = -2 * np.mean(y_train[idx] - y_hat)
                coef_der = -2 * np.dot((y_train[idx] - y_hat),X_train[idx])

                self.intercept_ = self.intercept_ - (self.lr * intercept_der)
                self.coef_ = self.coef_ - (self.lr * coef_der)

        print("Intercept :",self.intercept_)
        print("\nCoefficients :\n",self.coef_)

    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

## Using Sklearn :

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd = SGDRegressor(learning_rate='constant',eta0=0.1)
```

```python
batch_size = 35

for i in range(100):

    idx = random.sample(range(X_train.shape[0]),batch_size)
    sgd.partial_fit(X_train[idx],y_train[idx])  # Perform 1 epoch of Stochastic GD on given samples
```

```python
sgd.coef_
```

```python
sgd.intercept_
```

```python
y_pred = sgd.predict(X_test)
print("R2 Score :",r2_score(y_test,y_pred))
```

### NOTE :

- `Batch GD :` 1 update / epoch

- `Stochastic GD :` n updates / epoch

- `Mini Batch GD :`  batches_update / epoch

---
- For n rows , batch_size =n  :-
    - Only 1 Batch
    - Mini Batch GD = `Batch GD`
---
- For n rows , batch_size =1  :-
    - n Batches
    - Mini Batch GD = `Stochastic GD`
