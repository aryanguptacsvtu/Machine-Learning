## End of Distribution Imputation :

`CASE 1 : Column Data is Normally Distributed`

- Use (mean + 3.sigma) or (mean - 3.sigma) for Imputation


`CASE 2 : Column Data is Skewed`

- Use IQR Proximity rule for Imputation [Q1- 1.5*IQR] [Q3 + 1.5*IQR]

`PROS :`
- Easy to implement

`CONS :`
- Changes shape of Distribution
- May create extra Outliers
- Changes in Covariance / Correlation

### `When to USE :`

- When Data is `NOT` Missing Completely at Random (MCAR)
