```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('train.csv')[['Age','Pclass','SibSp','Parch','Survived']]
df.head(4)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
X = df.iloc[:,0:4]
y = df.iloc[:,-1]

X.sample(5)
```

```python
print("Cross Validation Score : ")
print(np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20)))
```

## Feature Construction :

Process of creating `New` and Useful `features` from existing data in order to improve Accurracy

```python
X['Family_size'] = X['SibSp'] + X['Parch'] + 1
X.head()
```

```python
def myfun(num):
    if (num == 1):
        return  0   #Alone
    elif(num>1 and num<=4):
        return 1    #Small Family
    else:
        return 2    #Large Family
```

```python
myfun(23) , myfun(3) , myfun(1)
```

```python
X['Family_type'] = X['Family_size'].apply(myfun)
X.sample(5)
```

```python
X.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
X.sample(5)
```

```python
print("Cross Validation Score : ")
print(np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20)))
```

## Feature Splitting :

It is the process of breaking down a single, complex feature into multiple, more informative `sub-features`.

```python
df2 = pd.read_csv('train.csv')
df2.head(2)
```

```python
df2['Name']
```

```python
# [1] => Content after (,)  ,  [0] => Content before (,)
df2['Name'].str.split(', ',expand=True)
```

```python
# [1] => Content after (.)  ,  [0] => Content before (.)
df2['Name'].str.split(', ',expand=True)[1].str.split('.',expand=True)
```

```python
df2['Title'] = df2['Name'].str.split(', ',expand=True)[1].str.split('.',expand=True)[0]
df2.sample(3)
```

```python
df2[['Title','Name']]
```

```python
(df2.groupby('Title')['Survived'].mean()).sort_values(ascending=False)
```

```python
df2['IsMarried'] = 0
df2['IsMarried'].loc[df2['Title'] == 'Mrs'] = 1
```
