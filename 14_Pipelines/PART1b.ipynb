```python
import numpy as np
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
ohe_sex = pickle.load(open('C:/Users/gupta/OneDrive/Documents/Machine Learning/14_Pipelines/ohe_sex.pkl','rb'))
ohe_embarked = pickle.load(open('C:/Users/gupta/OneDrive/Documents/Machine Learning/14_Pipelines/ohe_embarked.pkl','rb'))
clf = pickle.load(open('C:/Users/gupta/OneDrive/Documents/Machine Learning/14_Pipelines/clf.pkl','rb'))
```

```python
# Sample User Input 
# PClass/Gender/Age/SibSp/Parch/Fare/Embarked
test_input = np.array([2,'male',31.0, 0, 0, 10.5,'S'],dtype=object).reshape(1,7)
```

```python
test_input
```

### Applying Transformations :

```python
test_input_sex = ohe_sex.transform((test_input[:,1]).reshape(1,1))
test_input_sex
```

```python
test_input_embarked = ohe_embarked.transform((test_input[:,-1]).reshape(1,1))
test_input_embarked
```

```python
# Extracting 'age' column
test_input_age = test_input[:,2].reshape(1,1)
test_input_age
```

```python
# Joining
test_input_transformed = np.concatenate((test_input[:,[0,3,4,5]] , 
                                        test_input_age ,test_input_sex ,
                                        test_input_embarked),axis=1)
```

```python
test_input_transformed
```

### Without using Pipeline :

```python
ans = clf.predict(test_input_transformed)

if (ans == 1):
    print("Survived✅")
else:
    print("Not Survived❌")
```
